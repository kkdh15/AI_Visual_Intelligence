```python
def visualize(img):
    plt.figure(figsize=(10,10))
    plt.imshow(img, cmap='gray')
    width, height = img.shape
    for x in range(width):
        for y in range(height):
            plt.annotate(img[x,y],(y,x),horizontalalignment='center',verticalalignment='center',color='white')
```
